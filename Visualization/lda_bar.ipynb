```python
import numpy as np
import matplotlib.pyplot as plt

y = np.arange(5)
x = [49, 55, 54, 45, 60]
x2 = [25, 0, 0, 0, 0]
fig, ax = plt.subplots(figsize=(5,3))
ax.barh(y, x, height=.6)
ax.barh(y, x2, height=.6, left = x)
plt.text(s='Time', x=58, y=4, color="w", verticalalignment="center", horizontalalignment="right", size=14)
plt.text(s='Payment', x=43, y=3, color="w", verticalalignment="center", horizontalalignment="right", size=14)
plt.text(s='Refund', x=52, y=2, color="w", verticalalignment="center", horizontalalignment="right", size=14)
plt.text(s='Discount', x=53, y=1, color="w", verticalalignment="center", horizontalalignment="right", size=14)
plt.text(s='Ticket', x=46, y=0, color="w", verticalalignment="center", horizontalalignment="right", size=14)
plt.text(s='Frozen', x=72, y=0, color="w", verticalalignment="center", horizontalalignment="right", size=14)
plt.xticks([0,20,40,60,80])
plt.yticks([])
plt.ylabel('Topics',fontsize=30)
plt.tight_layout()
# plt.savefig('LDAtop.png')
plt.show()
```
