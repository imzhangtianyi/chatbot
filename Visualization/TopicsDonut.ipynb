```python
import matplotlib.pyplot as plt
 
# create data
names=['Discount Offer: 5.26%', 
       'Select Seat: 4.94%', 
       'Time: 4.73%', 
       'Payment: 4.59%', 
       'Children: 4.47%',
       'Refund: 2.93%',
       'Cashback: 1.43%',
       'Other: 71.7%']
size=[0.526,0.494,0.473,0.459,0.447,0.293,0.143,7.165]
 
# Create a circle for the center of the plot
my_circle=plt.Circle( (0,0), 0.7, color='white')

# Give color names
plt.figure(figsize=(6,6))
plt.pie(size, labels=names, 
        startangle=90,
        counterclock=False,
        colors=['#FF5733','#FF5733','#FF5733','#FF5733','#FF5733','#FF5733','#FF5733','skyblue'], 
        wedgeprops= {'edgecolor':'w', 'linewidth':3},
        textprops = {'fontsize':20})
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.axis('equal')
plt.show()
 

```
